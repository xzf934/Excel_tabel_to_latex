```python
import numpy as np
import pandas as pd
import glob
```

```python
folder_path = 'C:/research/india_outage/1214'
files = glob.glob(f'{folder_path}/byindustry*.xls')
```

```python
name_index = list()
for file in files:
    name_index.append([int(file.split('_')[-1].split('.')[0]),file])
name_index.sort(key=lambda x: x[0])
name_index
```

```python
df_summary = pd.DataFrame()

n_sheets = 8
sheets = pd.ExcelFile(name_index[0][1]).sheet_names

for sheet in sheets[0:n_sheets]:
    print(sheet)
    effect = list()
    se = list()
    pvalue = list()
    significance = list()
    for idx, file in name_index:
        if sheet == sheets[0]:
            print(idx, file)
        df = pd.read_excel(file, sheet_name=sheet, header=0, index_col=0)
        effect.append(df['b']['demeaned_shortage_percent'])
        se.append(df['se']['demeaned_shortage_percent'])
        pvalue.append(df['pvalue']['demeaned_shortage_percent'])
        pvalue_num = df['pvalue']['demeaned_shortage_percent']
        if pvalue_num < 0.001:
            significance.append('***')
        elif pvalue_num < 0.05:
            significance.append('**')
        elif pvalue_num < 0.1:
            significance.append('*')
        else:
            significance.append('')
    df_summary = pd.concat([df_summary,pd.DataFrame({f'effect_{sheet}':effect, f'se_{sheet}':se, f'pvalue_{sheet}':pvalue, f'sg_{sheet}':significance})], axis=1)
df_summary['category'] = pd.Categorical(df_summary.index+1)
df_summary
```
